```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
can_21 = "uv-canberra-2021.csv"
can_20 = "uv-canberra-2020.csv"
can_19 = "uv-canberra-2019.csv"
can_18 = "uv-canberra-2018.csv"
can_17 = "uv-canberra-2017.csv"

can_21_df = pd.read_csv(can_21, encoding="UTF-8")
can_20_df = pd.read_csv(can_20, encoding="UTF-8")
can_19_df = pd.read_csv(can_19, encoding="UTF-8")
can_18_df = pd.read_csv(can_18, encoding="UTF-8")
can_17_df = pd.read_csv(can_17, encoding="UTF-8")
```

```python
#Removing Lat and Lon for Canberra cvs. Splitting date to time, year, month and day
#This is done for each of the CVS years
can_17_df.drop(['Lat',"Lon"], axis=1, inplace=True)
can_17_df[['Date','Time']] = can_17_df['Date-Time'].str.split(' ', n=1, expand=True)
can_17_df[["year", "month", "day"]] = can_17_df["Date"].str.split("-", expand = True)
can_17_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
can_17_df
```

```python
can_18_df.drop(['Lat',"Lon"], axis=1, inplace=True)
can_18_df[['Date','Time']] = can_18_df['Date-Time'].str.split(' ', n=1, expand=True)
can_18_df[["year", "month", "day"]] = can_18_df["Date"].str.split("-", expand = True)
can_18_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
can_18_df
```

```python
can_19_df.drop(['Lat',"Lon"], axis=1, inplace=True)
can_19_df[['Date','Time']] = can_19_df['Date-Time'].str.split(' ', n=1, expand=True)
can_19_df[["year", "month", "day"]] = can_19_df["Date"].str.split("-", expand = True)
can_19_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
can_19_df
```

```python
can_20_df.drop(['Lat',"Lon"], axis=1, inplace=True)
can_20_df[['Date','Time']] = can_20_df['Date-Time'].str.split(' ', n=1, expand=True)
can_20_df[["year", "month", "day"]] = can_20_df["Date"].str.split("-", expand = True)
can_20_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
can_20_df
```

```python
can_21_df.drop(['Lat',"Lon"], axis=1, inplace=True)
can_21_df[['Date','Time']] = can_21_df['Date-Time'].str.split(' ', n=1, expand=True)
can_21_df[["year", "month", "day"]] = can_21_df["Date"].str.split("-", expand = True)
can_21_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
can_21_df
```

```python
#Obtaining UV index average per month for 2017
#Data time taken between 8am to 8pm to remove outliers of nighttime data
#As Canberra has daylight savings, data up to 8pm used to capture daylight in the summer
can_17_df['hour'] = pd.to_datetime(can_17_df['Time'], format='%H:%M:%S').dt.hour
clean_can17 = can_17_df[(can_17_df['hour'] >= 8) & (can_17_df['hour'] <=20)]

month_uv17 = clean_can17.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv17 = pd.DataFrame(month_uv17)
monthmean_uv17
```

```python
can_18_df['hour'] = pd.to_datetime(can_18_df['Time'], format='%H:%M:%S').dt.hour
clean_can18 = can_18_df[(can_18_df['hour'] >= 8) & (can_18_df['hour'] <=20)]

month_uv18 = clean_can18.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv18 = pd.DataFrame(month_uv18)
monthmean_uv18
```

```python
can_19_df['hour'] = pd.to_datetime(can_19_df['Time'], format='%H:%M:%S').dt.hour
clean_can19 = can_19_df[(can_19_df['hour'] >= 8) & (can_19_df['hour'] <=20)]

month_uv19 = clean_can19.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv19 = pd.DataFrame(month_uv19)
monthmean_uv19
```

```python
can_20_df['hour'] = pd.to_datetime(can_20_df['Time'], format='%H:%M:%S').dt.hour
clean_can20 = can_20_df[(can_20_df['hour'] >= 8) & (can_20_df['hour'] <=20)]

month_uv20 = clean_can20.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv20 = pd.DataFrame(month_uv20)
monthmean_uv20
```

```python
can_21_df['hour'] = pd.to_datetime(can_21_df['Time'], format='%H:%M:%S').dt.hour
clean_can21 = can_21_df[(can_21_df['hour'] >= 8) & (can_21_df['hour'] <=20)]

month_uv21 = clean_can21.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv21 = pd.DataFrame(month_uv21)
monthmean_uv21
```

```python
#Merging monthyly UV averages for each year together
merge_df = pd.merge(monthmean_uv17, monthmean_uv18, on="month", how="outer")

merge_df = merge_df.rename(columns={"UV_Index_x":"UV_mean_2017",
                                    "UV_Index_y":"UV_mean_2018"})

merge_df
```

```python
merge_df1 = pd.merge(merge_df, monthmean_uv19, on="month", how="outer")

merge_df1 = merge_df1.rename(columns={"UV_Index":"UV_mean_2019"})

merge_df1
```

```python
merge_df2 = pd.merge(merge_df1, monthmean_uv20, on="month", how="outer")

merge_df2 = merge_df2.rename(columns={"UV_Index":"UV_mean_2020"})

merge_df2
```

```python
merge_df3 = pd.merge(merge_df2, monthmean_uv21, on="month", how="outer")

merge_df3 = merge_df3.rename(columns={"UV_Index":"UV_mean_2021"})

merge_df3
```

```python
#Obtain mean of each month over the 5 years and output in mean column
merge_df3['mean'] = merge_df3.mean(axis=1)
canb_df = merge_df3
canb_df.to_csv("../Output/canb_df.csv", index=False, header=True)
canb_df
```

```python
#Plot graph and save graph to file
plt.plot(merge_df3['month'], merge_df3['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Month")
plt.ylabel("UV Index")
plt.title("Average Canberra UV index for last 5 years")
plt.savefig("../Images/CanberraUV.png")
```

```python
#Finding max UV index for each hour of day for years
hour_uv17 = can_17_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv18 = can_18_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv19 = can_19_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv20 = can_20_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv21 = can_21_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
```

```python
#Combining dataframes from each hour 
hour_merge = pd.merge(hour_uv17, hour_uv18, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index_x":"2017",
                                    "UV_Index_y":"2018"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv19, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2019"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv20, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2020"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv21, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2021"})

hour_merge
```

```python
#Produce mean column for max UV index scores for each hour of day
hour_merge['hour'] = hour_merge['hour'].astype(str)
hour_merge['mean'] = hour_merge.mean(axis=1)
can_hour = hour_merge
can_hour.to_csv("../Output/can_hour.csv", index=False, header=True)
can_hour
```

```python
#Plotting graph and saving graph image to Image folder
plt.plot(can_hour['hour'], can_hour['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Time of Day")
plt.ylabel("UV Index")
plt.title("Max UV index for time of day Canberra")
plt.savefig("../Images/CanHour.png")
```
