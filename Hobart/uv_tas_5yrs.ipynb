```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
tas_21 = "uv-kingston-2021.csv"
tas_20 = "uv-kingston-2020.csv"
tas_19 = "uv-kingston-2019.csv"
tas_18 = "uv-kingston-2018.csv"
tas_17 = "uv-kingston-2017.csv"

tas_21_df = pd.read_csv(tas_21, encoding="UTF-8")
tas_20_df = pd.read_csv(tas_20, encoding="UTF-8")
tas_19_df = pd.read_csv(tas_19, encoding="UTF-8")
tas_18_df = pd.read_csv(tas_18, encoding="UTF-8")
tas_17_df = pd.read_csv(tas_17, encoding="UTF-8")
```

```python
#Removing Lat and Lon for Kingtson cvs. Splitting date to time, year, month and day
#Kingston is a town 13km from Hobart
#This is done for each of the CVS years
tas_17_df.drop(['Lat',"Lon"], axis=1, inplace=True)
tas_17_df[['Date','Time']] = tas_17_df['Date-Time'].str.split(' ', n=1, expand=True)
tas_17_df[["year", "month", "day"]] = tas_17_df["Date"].str.split("-", expand = True)
tas_17_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
tas_17_df
```

```python
tas_18_df.drop(['Lat',"Lon"], axis=1, inplace=True)
tas_18_df[['Date','Time']] = tas_18_df['Date-Time'].str.split(' ', n=1, expand=True)
tas_18_df[["year", "month", "day"]] = tas_18_df["Date"].str.split("-", expand = True)
tas_18_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
tas_18_df
```

```python
tas_19_df.drop(['Lat',"Lon"], axis=1, inplace=True)
tas_19_df[['Date','Time']] = tas_19_df['Date-Time'].str.split(' ', n=1, expand=True)
tas_19_df[["year", "month", "day"]] = tas_19_df["Date"].str.split("-", expand = True)
tas_19_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
tas_19_df
```

```python
tas_20_df.drop(['Lat',"Lon"], axis=1, inplace=True)
tas_20_df[['Date','Time']] = tas_20_df['Date-Time'].str.split(' ', n=1, expand=True)
tas_20_df[["year", "month", "day"]] = tas_20_df["Date"].str.split("-", expand = True)
tas_20_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
tas_20_df
```

```python
tas_21_df.drop(['Lat',"Lon"], axis=1, inplace=True)
tas_21_df[['Date','Time']] = tas_21_df['Date-Time'].str.split(' ', n=1, expand=True)
tas_21_df[["year", "month", "day"]] = tas_21_df["Date"].str.split("-", expand = True)
tas_21_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
tas_21_df
```

```python
#Obtaining UV index average per month for 2017
#Data time taken between 8am to 8pm to remove outliers of nighttime data
#As Tasmania has daylight savings, data up to 8pm used to capture daylight in the summer
tas_17_df['hour'] = pd.to_datetime(tas_17_df['Time'], format='%H:%M:%S').dt.hour
clean_tas17 = tas_17_df[(tas_17_df['hour'] >= 8) & (tas_17_df['hour'] <=20)]

month_uv17 = clean_tas17.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv17 = pd.DataFrame(month_uv17)
monthmean_uv17
```

```python
tas_18_df['hour'] = pd.to_datetime(tas_18_df['Time'], format='%H:%M:%S').dt.hour
clean_tas18 = tas_18_df[(tas_18_df['hour'] >= 8) & (tas_18_df['hour'] <=20)]

month_uv18 = clean_tas18.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv18 = pd.DataFrame(month_uv18)
monthmean_uv18
```

```python
tas_19_df['hour'] = pd.to_datetime(tas_19_df['Time'], format='%H:%M:%S').dt.hour
clean_tas19 = tas_19_df[(tas_19_df['hour'] >= 8) & (tas_19_df['hour'] <=20)]

month_uv19 = clean_tas19.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv19 = pd.DataFrame(month_uv19)
monthmean_uv19
```

```python
tas_20_df['hour'] = pd.to_datetime(tas_20_df['Time'], format='%H:%M:%S').dt.hour
clean_tas20 = tas_20_df[(tas_20_df['hour'] >= 8) & (tas_20_df['hour'] <=20)]

month_uv20 = clean_tas20.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv20 = pd.DataFrame(month_uv20)
monthmean_uv20
```

```python
tas_21_df['hour'] = pd.to_datetime(tas_21_df['Time'], format='%H:%M:%S').dt.hour
clean_tas21 = tas_21_df[(tas_21_df['hour'] >= 8) & (tas_21_df['hour'] <=20)]

month_uv21 = clean_tas21.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv21 = pd.DataFrame(month_uv21)
monthmean_uv21
```

```python
#Merging monthyly UV averages for each year together
merge_df = pd.merge(monthmean_uv17, monthmean_uv18, on="month", how="outer")

merge_df = merge_df.rename(columns={"UV_Index_x":"UV_mean_2017",
                                    "UV_Index_y":"UV_mean_2018"})

merge_df
```

```python
merge_df1 = pd.merge(merge_df, monthmean_uv19, on="month", how="outer")

merge_df1 = merge_df1.rename(columns={"UV_Index":"UV_mean_2019"})

merge_df1
```

```python
merge_df2 = pd.merge(merge_df1, monthmean_uv20, on="month", how="outer")

merge_df2 = merge_df2.rename(columns={"UV_Index":"UV_mean_2020"})

merge_df2
```

```python
merge_df3 = pd.merge(merge_df2, monthmean_uv21, on="month", how="outer")

merge_df3 = merge_df3.rename(columns={"UV_Index":"UV_mean_2021"})

merge_df3
```

```python
#Obtain mean of each month over the 5 years and output in mean column
merge_df3['mean'] = merge_df3.mean(axis=1)
tas_df = merge_df3
tas_df.to_csv("../Output/tas_df.csv", index=False, header=True)
tas_df
```

```python
#Plot graph and save graph to file
plt.plot(merge_df3['month'], merge_df3['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Month")
plt.ylabel("UV Index")
plt.title("Average Kingston, Tasmania UV index for last 5 years")
plt.savefig("../Images/TasmaniaUV.png")
```

```python
#Finding max UV index for each hour of day for years
hour_uv17 = tas_17_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv18 = tas_18_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv19 = tas_19_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv20 = tas_20_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv21 = tas_21_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
```

```python
#Combining dataframes from each hour 
hour_merge = pd.merge(hour_uv17, hour_uv18, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index_x":"2017",
                                    "UV_Index_y":"2018"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv19, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2019"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv20, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2020"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv21, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2021"})

hour_merge
```

```python
#Produce mean column for max UV index scores for each hour of day
hour_merge['hour'] = hour_merge['hour'].astype(str)
hour_merge['mean'] = hour_merge.mean(axis=1)
tas_hour = hour_merge
tas_hour.to_csv("../Output/tas_hour.csv", index=False, header=True)
tas_hour
```

```python
#Plotting graph and saving graph image to Image folder
plt.plot(tas_hour['hour'], tas_hour['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Time of Day")
plt.ylabel("UV Index")
plt.title("Max UV index for time of day Tasmania")
plt.savefig("../Images/TasHour.png")
```
