```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
per_21 = "uv-perth-2021.csv"
per_20 = "uv-perth-2020.csv"
per_19 = "uv-perth-2019.csv"
per_18 = "uv-perth-2018.csv"
per_17 = "uv-perth-2017.csv"

per_21_df = pd.read_csv(per_21, encoding="UTF-8")
per_20_df = pd.read_csv(per_20, encoding="UTF-8")
per_19_df = pd.read_csv(per_19, encoding="UTF-8")
per_18_df = pd.read_csv(per_18, encoding="UTF-8")
per_17_df = pd.read_csv(per_17, encoding="UTF-8")
```

```python
#Removing Lat and Lon for Perth cvs. Splitting date to time, year, month and day
#This is done for each of the CVS years
per_17_df.drop(['Lat',"Lon"], axis=1, inplace=True)
per_17_df[['Date','Time']] = per_17_df['Date-Time'].str.split(' ', n=1, expand=True)
per_17_df[["year", "month", "day"]] = per_17_df["Date"].str.split("-", expand = True)
per_17_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
per_17_df
```

```python
per_18_df.drop(['Lat',"Lon"], axis=1, inplace=True)
per_18_df[['Date','Time']] = per_18_df['Date-Time'].str.split(' ', n=1, expand=True)
per_18_df[["year", "month", "day"]] = per_18_df["Date"].str.split("-", expand = True)
per_18_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
per_18_df
```

```python
per_19_df.drop(['Lat',"Lon"], axis=1, inplace=True)
per_19_df[['Date','Time']] = per_19_df['Date-Time'].str.split(' ', n=1, expand=True)
per_19_df[["year", "month", "day"]] = per_19_df["Date"].str.split("-", expand = True)
per_19_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
per_19_df
```

```python
per_20_df.drop(['Lat',"Lon"], axis=1, inplace=True)
per_20_df[['Date','Time']] = per_20_df['Date-Time'].str.split(' ', n=1, expand=True)
per_20_df[["year", "month", "day"]] = per_20_df["Date"].str.split("-", expand = True)
per_20_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
per_20_df
```

```python
per_21_df.drop(['Lat',"Lon"], axis=1, inplace=True)
per_21_df[['Date','Time']] = per_21_df['Date-Time'].str.split(' ', n=1, expand=True)
per_21_df[["year", "month", "day"]] = per_21_df["Date"].str.split("-", expand = True)
per_21_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
per_21_df
```

```python
#Obtaining UV index average per month for 2017
#Data time taken between 6am to 6pm to remove outliers of nighttime data
#As Darwin has no daylight savings, data from 6am to 6pm used
per_17_df['hour'] = pd.to_datetime(per_17_df['Time'], format='%H:%M:%S').dt.hour
clean_per17 = per_17_df[(per_17_df['hour'] >= 6) & (per_17_df['hour'] <=18)]

month_uv17 = clean_per17.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv17 = pd.DataFrame(month_uv17)
monthmean_uv17
```

```python
per_18_df['hour'] = pd.to_datetime(per_18_df['Time'], format='%H:%M:%S').dt.hour
clean_per18 = per_18_df[(per_18_df['hour'] >= 6) & (per_18_df['hour'] <=18)]

month_uv18 = clean_per18.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv18 = pd.DataFrame(month_uv18)
monthmean_uv18
```

```python
per_19_df['hour'] = pd.to_datetime(per_19_df['Time'], format='%H:%M:%S').dt.hour
clean_per19 = per_19_df[(per_19_df['hour'] >= 6) & (per_19_df['hour'] <=18)]

month_uv19 = clean_per19.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv19 = pd.DataFrame(month_uv19)
monthmean_uv19
```

```python
per_20_df['hour'] = pd.to_datetime(per_20_df['Time'], format='%H:%M:%S').dt.hour
clean_per20 = per_20_df[(per_20_df['hour'] >= 6) & (per_20_df['hour'] <=18)]

month_uv20 = clean_per20.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv20 = pd.DataFrame(month_uv20)
monthmean_uv20
```

```python
per_21_df['hour'] = pd.to_datetime(per_21_df['Time'], format='%H:%M:%S').dt.hour
clean_per21 = per_21_df[(per_21_df['hour'] >= 6) & (per_21_df['hour'] <=18)]

month_uv21 = clean_per21.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv21 = pd.DataFrame(month_uv21)
monthmean_uv21
```

```python
#Merging monthyly UV averages for each year together
merge_df = pd.merge(monthmean_uv17, monthmean_uv18, on="month", how="outer")

merge_df = merge_df.rename(columns={"UV_Index_x":"UV_mean_2017",
                                    "UV_Index_y":"UV_mean_2018"})

merge_df
```

```python
merge_df1 = pd.merge(merge_df, monthmean_uv19, on="month", how="outer")

merge_df1 = merge_df1.rename(columns={"UV_Index":"UV_mean_2019"})

merge_df1
```

```python
merge_df2 = pd.merge(merge_df1, monthmean_uv20, on="month", how="outer")

merge_df2 = merge_df2.rename(columns={"UV_Index":"UV_mean_2020"})

merge_df2
```

```python
merge_df3 = pd.merge(merge_df2, monthmean_uv21, on="month", how="outer")

merge_df3 = merge_df3.rename(columns={"UV_Index":"UV_mean_2021"})

merge_df3
```

```python
#Obtain mean of each month over the 5 years and output in mean column
merge_df3['mean'] = merge_df3.mean(axis=1)
per_df = merge_df3
per_df.to_csv("../Output/per_df.csv", index=False, header=True)
per_df
```

```python
#Plot graph and save graph to file
plt.plot(merge_df3['month'], merge_df3['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Month")
plt.ylabel("UV Index")
plt.title("Average Perth UV index for last 5 years")
plt.savefig("../Images/PerthUV.png")
```

```python
#Finding max UV index for each hour of day for years
hour_uv17 = per_17_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv18 = per_18_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv19 = per_19_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv20 = per_20_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv21 = per_21_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
```

```python
#Combining dataframes from each hour 
hour_merge = pd.merge(hour_uv17, hour_uv18, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index_x":"2017",
                                    "UV_Index_y":"2018"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv19, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2019"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv20, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2020"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv21, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2021"})

hour_merge
```

```python
#Produce mean column for max UV index scores for each hour of day
hour_merge['hour'] = hour_merge['hour'].astype(str)
hour_merge['mean'] = hour_merge.mean(axis=1)
perth_hour = hour_merge
perth_hour.to_csv("../Output/perth_hour.csv", index=False, header=True)
perth_hour
```

```python
#Plotting graph and saving graph image to Image folder
plt.plot(perth_hour['hour'], perth_hour['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Time of Day")
plt.ylabel("UV Index")
plt.title("Max UV index for time of day Perth")
plt.savefig("../Images/PerthHour.png")
```
