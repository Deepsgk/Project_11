```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
dar_21 = "uv-darwin-2021.csv"
dar_20 = "uv-darwin-2020.csv"
dar_19 = "uv-darwin-2019.csv"
dar_18 = "uv-darwin-2018.csv"
dar_17 = "uv-darwin-2017.csv"

dar_21_df = pd.read_csv(dar_21, encoding="UTF-8")
dar_20_df = pd.read_csv(dar_20, encoding="UTF-8")
dar_19_df = pd.read_csv(dar_19, encoding="UTF-8")
dar_18_df = pd.read_csv(dar_18, encoding="UTF-8")
dar_17_df = pd.read_csv(dar_17, encoding="UTF-8")
```

```python
#Removing Lat and Lon for Darwin cvs. Splitting date to time, year, month and day
#This is done for each of the CVS years
dar_17_df.drop(['Lat',"Lon"], axis=1, inplace=True)
dar_17_df[['Date','Time']] = dar_17_df['Date-Time'].str.split(' ', n=1, expand=True)
dar_17_df[["year", "month", "day"]] = dar_17_df["Date"].str.split("-", expand = True)
dar_17_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
dar_17_df
```

```python
dar_18_df.drop(['Lat',"Lon"], axis=1, inplace=True)
dar_18_df[['Date','Time']] = dar_18_df['Date-Time'].str.split(' ', n=1, expand=True)
dar_18_df[["year", "month", "day"]] = dar_18_df["Date"].str.split("-", expand = True)
dar_18_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
dar_18_df
```

```python
dar_19_df.drop(['Lat',"Lon"], axis=1, inplace=True)
dar_19_df[['Date','Time']] = dar_19_df['Date-Time'].str.split(' ', n=1, expand=True)
dar_19_df[["year", "month", "day"]] = dar_19_df["Date"].str.split("-", expand = True)
dar_19_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
dar_19_df
```

```python
dar_20_df.drop(['Lat',"Lon"], axis=1, inplace=True)
dar_20_df[['Date','Time']] = dar_20_df['Date-Time'].str.split(' ', n=1, expand=True)
dar_20_df[["year", "month", "day"]] = dar_20_df["Date"].str.split("-", expand = True)
dar_20_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
dar_20_df
```

```python
dar_21_df.drop(['Lat',"Lon"], axis=1, inplace=True)
dar_21_df[['Date','Time']] = dar_21_df['Date-Time'].str.split(' ', n=1, expand=True)
dar_21_df[["year", "month", "day"]] = dar_21_df["Date"].str.split("-", expand = True)
dar_21_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
dar_21_df
```

```python
#Obtaining UV index average per month for 2017
#Data time taken between 6am to 6pm to remove outliers of nighttime data
#As Darwin has no daylight savings, data from 6am to 6pm used
dar_17_df['hour'] = pd.to_datetime(dar_17_df['Time'], format='%H:%M:%S').dt.hour
clean_dar17 = dar_17_df[(dar_17_df['hour'] >= 6) & (dar_17_df['hour'] <=18)]

month_uv17 = clean_dar17.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv17 = pd.DataFrame(month_uv17)
monthmean_uv17
```

```python
dar_18_df['hour'] = pd.to_datetime(dar_18_df['Time'], format='%H:%M:%S').dt.hour
clean_dar18 = dar_18_df[(dar_18_df['hour'] >= 6) & (dar_18_df['hour'] <=18)]

month_uv18 = clean_dar18.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv18 = pd.DataFrame(month_uv18)
monthmean_uv18
```

```python
dar_19_df['hour'] = pd.to_datetime(dar_19_df['Time'], format='%H:%M:%S').dt.hour
clean_dar19 = dar_19_df[(dar_19_df['hour'] >= 6) & (dar_19_df['hour'] <=18)]

month_uv19 = clean_dar19.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv19 = pd.DataFrame(month_uv19)
monthmean_uv19
```

```python
dar_20_df['hour'] = pd.to_datetime(dar_20_df['Time'], format='%H:%M:%S').dt.hour
clean_dar20 = dar_20_df[(dar_20_df['hour'] >= 6) & (dar_20_df['hour'] <=18)]

month_uv20 = clean_dar20.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv20 = pd.DataFrame(month_uv20)
monthmean_uv20
```

```python
dar_21_df['hour'] = pd.to_datetime(dar_21_df['Time'], format='%H:%M:%S').dt.hour
clean_dar21 = dar_21_df[(dar_21_df['hour'] >= 6) & (dar_21_df['hour'] <=18)]

month_uv21 = clean_dar21.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv21 = pd.DataFrame(month_uv21)
monthmean_uv21
```

```python
#Merging monthyly UV averages for each year together
merge_df = pd.merge(monthmean_uv17, monthmean_uv18, on="month", how="outer")

merge_df = merge_df.rename(columns={"UV_Index_x":"UV_mean_2017",
                                    "UV_Index_y":"UV_mean_2018"})

merge_df
```

```python
merge_df1 = pd.merge(merge_df, monthmean_uv19, on="month", how="outer")

merge_df1 = merge_df1.rename(columns={"UV_Index":"UV_mean_2019"})

merge_df1
```

```python
merge_df2 = pd.merge(merge_df1, monthmean_uv20, on="month", how="outer")

merge_df2 = merge_df2.rename(columns={"UV_Index":"UV_mean_2020"})

merge_df2
```

```python
merge_df3 = pd.merge(merge_df2, monthmean_uv21, on="month", how="outer")

merge_df3 = merge_df3.rename(columns={"UV_Index":"UV_mean_2021"})

merge_df3
```

```python
#Obtain mean of each month over the 5 years and output in mean column
merge_df3['mean'] = merge_df3.mean(axis=1)
dar_df = merge_df3
dar_df.to_csv("../Output/dar_df.csv", index=False, header=True)
dar_df
```

```python
#Plot graph and save graph to file
plt.plot(merge_df3['month'], merge_df3['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Month")
plt.ylabel("UV Index")
plt.title("Average Darwin UV index for last 5 years")
plt.savefig("../Images/DarwinUV.png")
```

```python
#Finding max UV index for each hour of day for years
hour_uv17 = dar_17_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv18 = dar_18_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv19 = dar_19_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv20 = dar_20_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
hour_uv21 = dar_21_df.groupby('hour', as_index=False, sort=False)['UV_Index'].max()
```

```python
#Combining dataframes from each hour 
hour_merge = pd.merge(hour_uv17, hour_uv18, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index_x":"2017",
                                    "UV_Index_y":"2018"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv19, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2019"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv20, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2020"})

hour_merge
```

```python
hour_merge = pd.merge(hour_merge, hour_uv21, on="hour", how="outer")

hour_merge = hour_merge.rename(columns={"UV_Index":"2021"})

hour_merge
```

```python
#Produce mean column for max UV index scores for each hour of day
hour_merge['hour'] = hour_merge['hour'].astype(str)
hour_merge['mean'] = hour_merge.mean(axis=1)
darw_hour = hour_merge
darw_hour.to_csv("../Output/darw_hour.csv", index=False, header=True)
darw_hour
```

```python
#Plotting graph and saving graph image to Image folder
plt.plot(darw_hour['hour'], darw_hour['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("Time of Day")
plt.ylabel("UV Index")
plt.title("Max UV index for time of day Darwin")
plt.savefig("../Images/DarHour.png")
```
