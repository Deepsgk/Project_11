```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
ade_21 = "uv-adelaide-2021.csv"
ade_20 = "uv-adelaide-2020.csv"
ade_19 = "uv-adelaide-2019.csv"
ade_18 = "uv-adelaide-2018.csv"
ade_17 = "uv-adelaide-2017.csv"

ade_21_df = pd.read_csv(ade_21, encoding="UTF-8")
ade_20_df = pd.read_csv(ade_20, encoding="UTF-8")
ade_19_df = pd.read_csv(ade_19, encoding="UTF-8")
ade_18_df = pd.read_csv(ade_18, encoding="UTF-8")
ade_17_df = pd.read_csv(ade_17, encoding="UTF-8")
```

```python
#Removing Lat and Lon for Adelaide cvs. Splitting date to time, year, month and day
#This is done for each of the CVS years
ade_17_df.drop(['Lat',"Lon"], axis=1, inplace=True)
ade_17_df[['Date','Time']] = ade_17_df['Date-Time'].str.split(' ', n=1, expand=True)
ade_17_df[["year", "month", "day"]] = ade_17_df["Date"].str.split("-", expand = True)
ade_17_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
ade_17_df
```

```python
ade_18_df.drop(['Lat',"Lon"], axis=1, inplace=True)
ade_18_df[['Date','Time']] = ade_18_df['Date-Time'].str.split(' ', n=1, expand=True)
ade_18_df[["year", "month", "day"]] = ade_18_df["Date"].str.split("-", expand = True)
ade_18_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
ade_18_df
```

```python
ade_19_df.drop(['Lat',"Lon"], axis=1, inplace=True)
ade_19_df[['Date','Time']] = ade_19_df['Date-Time'].str.split(' ', n=1, expand=True)
ade_19_df[["year", "month", "day"]] = ade_19_df["Date"].str.split("-", expand = True)
ade_19_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
ade_19_df
```

```python
ade_20_df.drop(['Lat',"Lon"], axis=1, inplace=True)
ade_20_df[['Date','Time']] = ade_20_df['Date-Time'].str.split(' ', n=1, expand=True)
ade_20_df[["year", "month", "day"]] = ade_20_df["Date"].str.split("-", expand = True)
ade_20_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
ade_20_df
```

```python
ade_21_df.drop(['Lat',"Lon"], axis=1, inplace=True)
ade_21_df[['Date','Time']] = ade_21_df['Date-Time'].str.split(' ', n=1, expand=True)
ade_21_df[["year", "month", "day"]] = ade_21_df["Date"].str.split("-", expand = True)
ade_21_df.drop(['Date-Time', 'Date'], axis=1, inplace=True)
ade_21_df
```

```python
#Obtaining UV index average per month for 2017
#Data time taken between 8am to 8pm to remove outliers of nighttime data
#As Adelaide has daylight savings, data up to 8pm used to capture daylight in the summer
ade_17_df['hour'] = pd.to_datetime(ade_17_df['Time'], format='%H:%M:%S').dt.hour
clean_ade17 = ade_17_df[(ade_17_df['hour'] >= 8) & (ade_17_df['hour'] <=20)]

month_uv17 = clean_ade17.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv17 = pd.DataFrame(month_uv17)
monthmean_uv17
```

```python
ade_18_df['hour'] = pd.to_datetime(ade_18_df['Time'], format='%H:%M:%S').dt.hour
clean_ade18 = ade_18_df[(ade_18_df['hour'] >= 8) & (ade_18_df['hour'] <=20)]

month_uv18 = clean_ade18.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv18 = pd.DataFrame(month_uv18)
monthmean_uv18
```

```python
ade_19_df['hour'] = pd.to_datetime(ade_19_df['Time'], format='%H:%M:%S').dt.hour
clean_ade19 = ade_19_df[(ade_19_df['hour'] >= 8) & (ade_19_df['hour'] <=20)]

month_uv19 = clean_ade19.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv19 = pd.DataFrame(month_uv19)
monthmean_uv19
```

```python
ade_20_df['hour'] = pd.to_datetime(ade_20_df['Time'], format='%H:%M:%S').dt.hour
clean_ade20 = ade_20_df[(ade_20_df['hour'] >= 8) & (ade_20_df['hour'] <=20)]

month_uv20 = clean_ade20.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv20 = pd.DataFrame(month_uv20)
monthmean_uv20
```

```python
ade_21_df['hour'] = pd.to_datetime(ade_21_df['Time'], format='%H:%M:%S').dt.hour
clean_ade21 = ade_21_df[(ade_21_df['hour'] >= 8) & (ade_21_df['hour'] <=20)]

month_uv21 = clean_ade21.groupby('month', as_index=False, sort=False)['UV_Index'].mean()
monthmean_uv21 = pd.DataFrame(month_uv21)
monthmean_uv21
```

```python
#Merging monthyly UV averages for each year together
merge_df = pd.merge(monthmean_uv17, monthmean_uv18, on="month", how="outer")

merge_df = merge_df.rename(columns={"UV_Index_x":"UV_mean_2017",
                                    "UV_Index_y":"UV_mean_2018"})

merge_df
```

```python
merge_df1 = pd.merge(merge_df, monthmean_uv19, on="month", how="outer")

merge_df1 = merge_df1.rename(columns={"UV_Index":"UV_mean_2019"})

merge_df1
```

```python
merge_df2 = pd.merge(merge_df1, monthmean_uv20, on="month", how="outer")

merge_df2 = merge_df2.rename(columns={"UV_Index":"UV_mean_2020"})

merge_df2
```

```python
merge_df3 = pd.merge(merge_df2, monthmean_uv21, on="month", how="outer")

merge_df3 = merge_df3.rename(columns={"UV_Index":"UV_mean_2021"})

merge_df3
```

```python
#Obtain mean of each month over the 5 years and output in mean column
merge_df3['mean'] = merge_df3.mean(axis=1)
merge_df3
```

```python
plt.plot(merge_df3['month'], merge_df3['mean'], marker="+", color="blue", linewidth=1)
plt.xlabel("month")
plt.ylabel("mean")
plt.title("Average Brisbane UV index for last 5 years")
plt.savefig("BrisbaneUV.png")
```
